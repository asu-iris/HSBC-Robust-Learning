```python
import numpy as np
import sys,os
from matplotlib import pyplot as plt

sys.path.append(os.path.abspath(os.path.dirname(os.getcwd())))
sys.path.append(os.path.abspath(os.getcwd()))
```

```python
#walker
results_random =[]
for j in (0,1,2,3,4):
    res = np.load('../Data/Ablation/disagreement/0/run_{}/eval_result_0.npz'.format(j))
    results=[]
    for i in range(0,81,5):
        results.append(res[str(i)])
    results_random.append(results)
std_random = np.std(results_random,axis=0)
results_random = np.mean(results_random,axis=0)

results_50 =[]
for j in (0,1,2,3,4):
    res = np.load('../Data/Ablation/disagreement/50/run_{}/eval_result_0.npz'.format(j))
    results=[]
    for i in range(0,81,5):
        results.append(res[str(i)])
    results_50.append(results)
std_50 = np.std(results_50,axis=0)
results_50 = np.mean(results_50,axis=0)

results_75 =[]
for j in (0,1,2,3,4):
    res = np.load('../Data/Ablation/disagreement/75/run_{}/eval_result_0.npz'.format(j))
    results=[]
    for i in range(0,81,5):
        results.append(res[str(i)])
    results_75.append(results)
std_75 = np.std(results_75,axis=0)
results_75 = np.mean(results_75,axis=0)
```

```python
#plt.title("Walker Abalation: Disagreement Threshold")
plt.grid()
plt.plot(np.arange(0,82,5)*10,results_75,label='$\eta$ = 0.75')
plt.fill_between(np.arange(0,82,5)*10,results_75-std_75,results_75+std_75,alpha=0.2)
plt.plot(np.arange(0,82,5)*10,results_50,label='$\eta$ = 0.50')
plt.fill_between(np.arange(0,82,5)*10,results_50-std_50,results_50+std_50,alpha=0.2)
plt.plot(np.arange(0,82,5)*10,results_random,label='$\eta$ = 0.0')
plt.fill_between(np.arange(0,82,5)*10,results_random-std_random,results_random+std_random,alpha=0.2)
plt.ylim(0,500)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel("Query Number",fontsize = 16)
plt.ylabel("Reward Sum",fontsize = 16)
plt.legend(fontsize=16)


plt.savefig('../Data/Figs/ablation/disagree.jpg',bbox_inches="tight")
```
